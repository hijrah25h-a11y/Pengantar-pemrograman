```python
import pandas as pd #memanggil pandas ke libray untuk membaca file excel
from scipy.stats.mstats import gmean, hmean #untuk analisis data
import matplotlib.pyplot as plt #untuk membuat grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') #menyimpan sebuah data excel dalam variabel my_dataset

a_mean = my_dataset.Zr.mean() #menghitung rata-rata aritmatika dari kolom Zr
g_mean = gmean(my_dataset['Zr']) #menghitung rata-rata geomrtri dari data Zr
h_mean = hmean(my_dataset['Zr']) #menghitung rata-rata harmonik dari data Zr

print ('-------') #mencetak garis pemisah di output
print ('arithmetic mean') #menampilkan judul 
print ("{0:.1f} [ppm]".format(a_mean)) #untuk menampilkan nilai a_mean dengan 1 angka di belakang koma + satuan ppm
print ('-------') #garis pemisah terakhir

print ('geometric mean') #menampilkan judul
print ("{0:.1f} [ppm]".format(g_mean)) #menampilkan nilai g_mean
print ('-------') #garis pemisah 

print ('harmonic mean') #menampilakan judul
print ("{0:.1f} [ppm]".format(h_mean)) #menampilkan mnilai h_mean
print ('-------') #garis pemisah terakhir

fig, ax = plt.subplots() #fig untuk menggambar seluruh gambar grafik, ax untuk menggambar area grafik
ax.hist(my_dataset.Zr, bins='auto', density=True, edgecolor='k', label='Measurements Hist', alpha=0.8) #membuat histogram atau diagram batang distribusi data dari nilai Zr
ax.axvline(a_mean, color='red', label='Arithmetic mean', linewidth=2) #memberi warna ungu pada area aritmetic mean
ax.axvline(g_mean, color='green',  label='Geometric mean', linewidth=2) #memberi warna oren pada area geometric mean
ax.axvline(h_mean, color='pink',  label='Harmonic mean', linewidth=2) #memberi warna hijau pada area harmonic mean
ax.set_xlabel('Zr [ppm]') #memberi nama pada sumbu x
ax.set_ylabel('Probability density') #meberi nama pada sumbu y
ax.legend() #untuk menampilkan legenda pada grafik

'''
Output:
-------
arithmetic mean
365.4 [ppm]
-------
geometric mean
348.6 [ppm]
-------
harmonic mean
333.8 [ppm]
-------
'''
```

```python
import pandas as pd #memanggil pandas untuk membaca file excel
import numpy as np #untuk distribusi data
import matplotlib.pyplot as plt #untuk membuat grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') #menyimpan data excel ke dalam variabel

hist, bin_edges = np.histogram(my_dataset['Zr'], bins= 20, density=True)
# np.histogram(...) = menghitung distribusi data tanpa menampilkan grafik
# my_dataset['Zr'] = mengambil data dari kolom Zr
# bins=20 = membagi data menjadi 20 kelas interval
# density=True = hasil histogram dalam bentuk probabilitas (bukan jumlah)
# hist = nilai tinggi tiap bin (frekuensi / densitas)
# bin_edges = batas-batas tiap bin (array panjangnya bins+1)
modal_value = (bin_edges[hist.argmax()] +  bin_edges[hist.argmax()+1])/2
# hist.argmax() = mencari indeks bin dengan nilai tertinggi (modus)
# bin_edges[hist.argmax()] = batas bawah bin modus
# bin_edges[hist.argmax()+1] = batas atas bin modus
# dijumlah lalu dibagi 2 = titik tengah bin → ini dianggap sebagai nilai modus

print ('modal value: {0:.0f} [ppm]'.format(modal_value))
# print(...) = menampilkan hasil ke layar
# {0:.0f} = format angka tanpa desimal (dibulatkan)
# [ppm] = satuan (parts per million)
fig, ax = plt.subplots() 
# plt.subplots() = membuat "kanvas" dan "area plot"
# fig (figure) = seluruh halaman / kanvas gambar
# ax (axes) = area tempat grafik digambar

ax.hist(my_dataset.Zr, bins=20, density=True, edgecolor='k', label="Measurements Hist", alpha=0.8) 
# ax.hist(...) = membuat histogram (diagram batang distribusi data)
# my_dataset.Zr = mengambil data dari kolom "Zr" di dataframe
# bins=20 = jumlah pembagian batang (20 kelas interval)
# density=True = mengubah histogram jadi bentuk probabilitas (bukan jumlah data)
# edgecolor='k' = warna garis tepi batang (k = hitam)
# label="Measurements Hist" = label untuk legenda
# alpha=0.8 = transparansi (0 = transparan, 1 = solid)

ax.axvline(modal_value, color="black", label="Modal value", linewidth=2)
# ax.axvline(...) = membuat garis vertikal pada grafik
# median = posisi garis di nilai median data
# color='orange' = warna garis
# label='Median' = label untuk legenda
# linewidth=2 = ketebalan garis

ax.set_xlabel('Zr [ppm]') #memberi nama pada sumbu x
ax.set_ylabel('Probability density') #memberi nama pada sumbu y
ax.legend() #untuk menampilkan legenda pada grafik

'''
Output: modal value: 277 [ppm]
'''
```

```python
import pandas as pd #untuk membaca  file excel
import matplotlib.pyplot as plt #untuk membuat grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') #menyimpan data file excel ke variabel my_dataset

variance =  my_dataset['Zr'].var()
# menghitung variance (ragam) dari data Zr
# menunjukkan seberapa besar penyebaran data terhadap
stddev =  my_dataset['Zr'].std()
# menghitung standar deviasi dari data Zr
# menunjukkan seberapa jauh data menyebar dari rata-rata (akar dari variance)

print ('-------') #menampilkan garis pemisah di output
print ('Variance') #menampilkan nama variance sebagai judul
print("{0:.0f} [square ppm]".format(variance)) #menampilakn nilai variance
print ('-------') #menampilkan garis pemisah
print ('Standard Deviation') #menampilakan judul standard deviation
print("{0:.0f} [ppm]".format(stddev)) #menampilkan nilai stddev
print ('-------') #menampilkan garis pemisah

fig, ax = plt.subplots() #fig untuk menampilkan kanvas, ax untuk menampilkan area grafik
ax.hist(my_dataset.Zr, bins= 20, density = True, edgecolor='k', label='Measurements Hist') #menampilkan histogram dari file dalam variabel my_dataset
ax.axvline(my_dataset['Zr'].mean()-stddev, color='pink', label=r'mean - 1$\sigma$', linewidth=2) #memberi warna pink pada mean -1
ax.axvline(my_dataset['Zr'].mean()+stddev, color='orange', label=r'mean + 1$\sigma$', linewidth=2) #memberi warna orange pada mean +1
ax.axvspan(my_dataset['Zr'].mean()-stddev, my_dataset['Zr'].mean()+stddev, alpha=0.1, color='blue', label=r'mean $\pm$ 1$\sigma$') #memberi warna blue pada  mean -+ 1
ax.set_xlabel('Zr [ppm]') #memberi nama pada sumbu x
ax.set_ylabel('Probability density') #memberi nama pada sumbu y
ax.legend() #menampilkan legenda pada grafik

'''
Output:
-------
Variance
14021 [square ppm]
-------
Standard Deviation
118 [ppm]
-------
'''
```

```python
import pandas as pd #membaca file excel
import numpy as np #untuk mengelola data angka
import matplotlib.pyplot as plt #untuk menampilkan grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') #meyimpan data excel ke dalam variabel my_dataset

q1 = np.percentile(my_dataset.Zr, 25, method = 'midpoint') #menghitung kuartil 1 (25%) dari data Zr batas bawah IQR
q3 = np.percentile(my_dataset.Zr, 75, method = 'midpoint') #menghitung kuartil 3 (75%) dari data Zr batas atas IQR
  
iqr = q3 - q1 # Interquaritle range (IQR)  

print ('-------') #menampilkan garis pemisah di output
print ('Interquaritle range (IQR): {0:.0f} [ppm]'.format(iqr)) #untuk menampilkan interquaritle range beserta nilai nya tanpa desimal
print ('-------') #menampilkan garis pemisah

fig, ax = plt.subplots() #fig untuk menampilkan kanvas, ax untuk menampilkan area grafik
ax.hist(my_dataset.Zr, bins='auto', density=True, edgecolor='k', label='Measurements Hist') #menampilkan histogram dari variabel my_dataset
ax.axvline(q1, color='pink', label='Q1', linewidth=2) #menapilkan warna pink pada Q1
ax.axvline(q3, color='black', label='Q3', linewidth=2) #menapilkan warna black pada Q3
ax.axvspan(q1, q3, alpha=0.1, color='orange', label='Interquaritle range (IQR)') #untuk menandai area pada grafik antara nilai Q1 dan Q3 yaitu rentang IQR
ax.set_xlabel('Zr [ppm]') #memberi nama pada sumbu x
ax.set_ylabel('Probability density') #memberi nama pada sumbu y
ax.legend() #menampilkan legenda pada grafik

'''
Output:
-------
Interquaritle range (IQR): 164 [ppm]
-------
'''
```

```python
import numpy as np #untuk perhitungan matematika
from scipy.stats import skew #untuk menghitung nilai skewness

a_mean = my_dataset.Zr.mean() #menghitung nilai rata-rata dari kolom Zr
median = my_dataset.Zr.median() #menghitung nilai tengah dari kolom Zr
hist, bin_edges = np.histogram(my_dataset['Zr'], bins=20, density=True) 
# membuat histogram dalam bentuk data:
# hist = tinggi/frekuensi tiap bin
# bin_edges = batas-batas tiap bin
modal_value = (bin_edges[hist.argmax()] + bin_edges[hist.argmax()+1])/2
# mencari nilai modus (nilai paling sering muncul):
# hist.argmax() = indeks bin dengan frekuensi tertinggi
# lalu diambil titik tengah dari bin tersebut
standard_deviation = my_dataset['Zr'].std() #menghitung standar deviasi (sebaran data dari rata-rata)

a1 = (a_mean - modal_value) / standard_deviation 
# Pearson skewness pertama:
# membandingkan mean dengan modus
a2 = 3 * (a_mean - median) / standard_deviation
# Pearson skewness kedua:
# membandingkan mean dengan median
g1 = skew(my_dataset['Zr']) #menghitung skewness menggunakan metode Fisher-Pearson (lebih umum dipakai)


print ('-------') #membuat garis pemisah
print ("Pearson's first coefficient of skewness: {:.2f}".format(a1)) #menampilkan nilai skewness Pearson pertama (2 angka di belakang koma)
print ("Pearson's 2nd moment of skewness: {:.2f}".format(a2)) #menampilkan nilai skewness Pearson kedua
print ("Fisher-Pearson's coefficient of skewness: {:.2f}".format(g1)) # menampilkan skewness versi scipy (Fisher-Pearson)
print ('-------') #garis penutup

'''
Output:
-------
Pearson's first coefficient of skewness: 0.74
Pearson's 2nd moment of skewness: 0.66
Fisher-Pearson's coefficient of skewness: 1.26
-------
'''
```
